**3. Create a new column in the dataframe "age_group". This column should contain the age group the respondent belongs to. The age group are 18-25,25-40,40-55,55+.**

```python
data=survey.copy()    # To create a copy of dataframe
```

```python
data.head(2)
```

```python
data.insert(10,'age_group',survey.age)   # To insert new column
```

```python
data.head(2)      
```

```python
data1=data[(data.age_group >= 18) & (data.age_group < 25)]
```

```python
data1.age_group.unique()
```

```python
 data1['age_group'] = '18-25'                              #Assign value to column
```

```python
data1.head(4)
```

```python
data2 = data[(data.age_group >= 25) & (data.age_group < 40)]
```

```python
data2['age_group']= '25-40'
```

```python
data2.head(4)
```

```python
data3 = data[(data.age_group >= 40)&(data.age_group < 55)]
```

```python
data3['age_group'] = '40-55'
```

```python
data3.head(4)
```

```python
data4 = data[data.age_group > 55]
```

```python
data4['age_group'] = '55+'
```

```python
data4.head(4)
```

```python
#Now we will do concatenation of dataframes.

data=pd.concat([data1,data2,data3,data4])
```
